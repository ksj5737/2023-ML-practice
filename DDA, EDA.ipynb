```python
df1.pivot_table(index='메세지', values='정지시간2', aggfunc = 'mean')  
#index = '구분하고 싶은 항목(범주형)'
#values = '계산하고자 하는 값(연속형)'
#aggfunc = '계산하고자 하는 통계량'
```

```python
#오류 메세지 '메세지' 항목에서 'Dodr Open'에 해당하는 오류들 중에 '정지시간2'이 가장 오래 걸린 상위 100개 데이터를 추출

cond3 = df1['메세지'] == 'Door Open'
df1.loc[cond3].sort_values(by = "정지시간2", ascending = False).head(100)
```

```python
#'정지시간2' 이 300초 이상이면서 '작업자'가 '이은미' 인 데이터를 추출해 '정지시간2' 의 요약통계량을 확인

cond4 = df1['정지시간2'] >= 300
cond5 = df1['작업자'] == '이은미'
df1.loc[cond4 & cond5].describe()

```
