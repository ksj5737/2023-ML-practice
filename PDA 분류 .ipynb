```python
#'Target', 'Error_message', 'target_binom', 'Chamber_Route' 컬럼은 공정 중에 발생한 것이 아닌 공정이 끝나고 품질 검사후 나온 데이터
# -> 즉 위 4개 컬럼은 독립변수(x)가 아니고 종속변수(y)이다.
df1.columns
```

```python
Y = df1['target_binom']

#drop() : 데이터 프레임에서 불필요한 항목을 제거

X = df1.drop(columns=['Target', 'Error_message', 'target_binom', 'Chamber_Route', 'Ox_Chamber', 'process', 'type',
                     'No_Die',  'Unnamed: 0_x', 'photo_soft_Chamber', 'process 2', 'Unnamed: 0_y', 'Etching_Chamber', 
                      'Process 3', 'Chamber_Num', 'process4', 'Unnamed: 0'])
```

```python
# 머신러닝에 불필요한 문자형 데이터가 포함되어있는 것을 확인

X.info()
```

```python
#숫자데이터만 선택하고 list로 만듬
numeric_list = X.describe().columns.tolist()
```

```python
X1 = X[numeric_list]
```

```python
#학습데이터와 검증데이터 분할 

from sklearn.model_selection import train_test_split
```

```python
# X, Y 학습, 검증데이터 각각 나눠줌, 기본값은 8:2 비율 
# 사이즈 비율 변경 : X_train, X_test, Y_train, Y_test = train_test_split(X1,Y, test_size=0.3)

X_train, X_test, Y_train, Y_test = train_test_split(X1,Y, random_state=1234)
#random_state  무작위 추출을 원하는 index로 맞추어 추출 -> 같은 데이터를 사용하니 사람마다 검증 결과가 달라지지 않고 같아진다

print(X_train.shape)  #shape : 데이터의 행과 열개수를 출력 
print(X_test.shape)
```

```python
#학습 수행
from sklearn.tree import DecisionTreeClassifier 
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
# 평가 방법
# - 학습 능력 : Y_train(실제값) - Y_train_pred(예측값) 비교
# - 일반화 능력 : Y_test(실제값) - Y_test_pred(예측값) 비교 

from sklearn.metrics import accuracy_score
```

```python
Y_train_pred = model.predict(X_train)
```

```python
#정확도 : 전체 데이터 중 정확하게 분류한 데이터의 수 0~1 사이값 

#학습능력 평가, 결과 =100%
accuracy_score(Y_train, Y_train_pred)
```

```python
Y_test_pred = model.predict(X_test)


#일반화 능력 평가 = 87%
accuracy_score(Y_test, Y_test_pred)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
precision_score(Y_test, Y_test_pred)
```

```python
recall_score(Y_test,Y_test_pred)
```

```python
f1_score(Y_test,Y_test_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test, Y_test_pred))
```
